```python
def substract(v1, v2):
    return (v1[0]-v2[0], v1[1]-v2[1])
```

```python
def distance(v1, v2):
    return length(substract(v1, v2))
```

```python
def perimeter(vectors):
    return sum([distance(vectors[i], vectors[i-1]) for i in range(0, len(vectors))])
```

```python
def to_polar(vector):
    x, y = vector[0], vector[1]
    angle = atan2(y, x)
    return (length(vector), angle)
```

```python
rotation_angle = pi/4
dino_polar = [to_polar(v) for v in dino_vectors]
dino_rotated_polar = [(l, angle+rotation_angle) for l,angle in dino_polar]
dino_rotated = [to_cartession(p) for p in dino_rotated_polar]
draw(
    Polygon(*dino_vectors, color=gray),
    Polygon(*dino_rotated, color=red)
)
```

```python
def rotate(angle, vectors):
    polar_vectors = [to_polar(v) for v in vectors]
    return [to_cartession((l, a+angle)) for l, a in polar_vectors]
```

```python
new_dino = translate((8,8), rotate(5*pi/3, dino_vectors))
```

```python
draw(
    Polygon(*dino_vectors, color=gray),
    Polygon(*new_dino, color=red)
)
```
